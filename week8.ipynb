```python
#  1. Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: For interactive choropleth maps
# import plotly.express as px

# Set style for plots
sns.set(style="darkgrid")
plt.rcParams["figure.figsize"] = (12, 6)
#  2. Load the COVID-19 Dataset
df = pd.read_csv("owid-covid-data.csv")

# Preview the first few rows
df.head()
# Convert 'date' to datetime format
df['date'] = pd.to_datetime(df['date'])


# Filter countries of interest
countries = ['Kenya', 'United States', 'India']
df = df[df['location'].isin(countries)]


# Drop rows with missing critical values
df = df.dropna(subset=['date', 'total_cases', 'total_deaths'])


# Fill other missing numeric values forward
df = df.ffill()



# Add death rate column
df['death_rate'] = df['total_deaths'] / df['total_cases']


df.head()

## visualization
plt.figure(figsize=(12, 6))
for country in countries:
    country_data = df[df['location'] == country]
    plt.plot(country_data['date'], country_data['total_cases'], label=country)

plt.title("Total COVID-19 Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.legend()
plt.show()


##Ttotalcdeaths over time
plt.figure(figsize=(12, 6))
for country in countries:
    country_data = df[df['location'] == country]
    plt.plot(country_data['date'], country_data['total_deaths'], label=country)

plt.title("Total COVID-19 Deaths Over Time")
plt.xlabel("Date")
plt.ylabel("Total Deaths")
plt.legend()
plt.show()


## vacinnation time
plt.figure(figsize=(12, 6))
for country in countries:
    country_data = df[df['location'] == country]
    plt.plot(country_data['date'], country_data['people_vaccinated'], label=country)

plt.title("Cumulative People Vaccinated Over Time")
plt.xlabel("Date")
plt.ylabel("People Vaccinated")
plt.legend()
plt.show()



```

##  Key Insights from the COVID-19 Data

-  **The United States** had the highest total cases and deaths overall.
-  **India** experienced a sharp rise in cases in mid-2021.
-  **Kenya's** vaccine rollout was slower compared to the US and India.
-  The **death rate** appears lower in India, possibly due to underreporting or age distribution.
